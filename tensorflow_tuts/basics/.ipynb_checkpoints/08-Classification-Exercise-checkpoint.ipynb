### THE DATA

** Read in the census_data.csv data with pandas**

```python
import pandas as pd
```

```python
census = pd.read_csv("census_data.csv")
```

```python
census.head()
```

** TensorFlow won't be able to understand strings as labels, you'll need to use pandas .apply() method to apply a custom function that converts them to 0s and 1s. This might be hard if you aren't very familiar with pandas, so feel free to take a peek at the solutions for this part.**

** Convert the Label column to 0s and 1s instead of strings.**

```python
#convert target into binary value of 0 and 1 labels
census.income_bracket.unique()
```

```python
def target_conv(row):
    if row ==" <=50K":
        return 0
    return 1

census.income_bracket = census.income_bracket.apply(target_conv)
```

```python
census.income_bracket.value_counts()
```

```python
census.head()
```

### Perform a Train Test Split on the Data

```python
from sklearn.model_selection import train_test_split
```

```python
y = census.income_bracket
x = census.drop("income_bracket",axis=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=101)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

### Create the Feature Columns for tf.esitmator

** Take note of categorical vs continuous values! **

```python
#recall we have both numerical and categorical values
features = census.columns.tolist()
features.remove("income_bracket")
```

```python
features
```

** Import Tensorflow **

```python
import tensorflow as tf
```

** Create the tf.feature_columns for the categorical values. Use vocabulary lists or just use hash buckets. **

```python
#for categorical first, using hash buckets because we don't want to bother exploring the individual category names
#becareful if the max bucket_size will run out

gender = tf.feature_column.categorical_column_with_hash_bucket("gender",hash_bucket_size=1000)
occupation = tf.feature_column.categorical_column_with_hash_bucket("occupation",hash_bucket_size=1000)
marital_status = tf.feature_column.categorical_column_with_hash_bucket("marital_status",hash_bucket_size=1000)
relationship = tf.feature_column.categorical_column_with_hash_bucket("relationship",hash_bucket_size=1000)
education = tf.feature_column.categorical_column_with_hash_bucket("education",hash_bucket_size=1000)
workclass = tf.feature_column.categorical_column_with_hash_bucket("workclass",hash_bucket_size=1000)
native_country = tf.feature_column.categorical_column_with_hash_bucket("native_country",hash_bucket_size=1000)
```

** Create the continuous feature_columns for the continuous values using numeric_column **

```python
age = tf.feature_column.numeric_column("age")
education_num = tf.feature_column.numeric_column("education_num")
capital_gain = tf.feature_column.numeric_column("capital_gain")
capital_loss = tf.feature_column.numeric_column("capital_loss")
hours_per_week = tf.feature_column.numeric_column("hours_per_week")
```

** Put all these variables into a single list with the variable name feat_cols **

```python
#assemble both types togeter
feat_cols = [gender,occupation, marital_status, relationship, education, workclass, native_country, age, education_num,
            capital_gain, capital_loss, hours_per_week]
```

### Create Input Function

** Batch_size is up to you. But do make sure to shuffle!**

```python
#we have not normalized with MinMax(), we'll be using LinearClassifier, not DNN
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=100,num_epochs=None,shuffle=True)
```

#### Create your model with tf.estimator

**Create a LinearClassifier.(If you want to use a DNNClassifier, keep in mind you'll need to create embedded columns out of the cateogrical feature that use strings, check out the previous lecture on this for more info.)**

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols)
```

** Train your model on the data, for at least 5000 steps. **

```python
model.train(input_fn=input_func,steps=5000)
```

### Evaluation

** Create a prediction input function. Remember to only supprt X_test data and keep shuffle=False. **

```python
input_test = tf.estimator.inputs.pandas_input_fn(x=X_test,shuffle=False)
```

** Use model.predict() and pass in your input function. This will produce a generator of predictions, which you can then transform into a list, with list() **

```python
pred_ = model.predict(input_fn=input_test)
```

** Each item in your list will look like this: **

```python
pred = list(pred_)
pred
```

** Create a list of only the class_ids key values from the prediction list of dictionaries, these are the predictions you will use to compare against the real y_test values. **

```python
final_preds = [i["class_ids"][0] for i in pred]
final_preds[0:10]
```

** Import classification_report from sklearn.metrics and then see if you can figure out how to use it to easily get a full report of your model's performance on the test data. **

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, final_preds))
```
