```python
import pandas as pd
import numpy as np
```

```python
loans_2007 = pd.read_csv('loans_2007.csv')
loans_2007.drop_duplicates(inplace=True)
loans_2007.head()
```

```python
loans_2007.shape
```

**Columns that contains future information or leakage of information, as well as intangible meaning (ie: ID) that should be removed. Below are such examples.**

**id**: id purpose.

**member_id**: id purpose.

**funded_amnt**: future leakage (decided to be funded).

**funded_amnt_inv**: by investor, same as above. Future leakage (decided to be funded).

**grade**: overlapping with interest rate, we don't want double significance.

**sub_grade**: also overlapping data with interest rate.

**emp_title**: salary will take care of this. Painful for categorical.

**issue_d**: leak from the future, the month of which it was issued.

**zip_code**: not practical. We have state data.

**out_prncp**: outstanding principle, already started to accept the loan request.

**out_prncp_inv**: by investor, same as above, already accepted the loan request.

**total_pymnt**: data leakage, the loan has been paid.

**total_pymnt_inv**: by investor, same as above - leakage.

**total_rec_prncp**: principle received to date - leakage.

**total_rec_int**: interest received 'till now - leakage.

**total_rec_late_fee**: late fee receive 'till now - leakage

**recoveries**: charge off gross recoveries - leakage

**collection_recovery_fee**: charge off collection fee - leakage

**last_pymnt_d**: last month of payment - leakage

**last_pymnt_amnt**: last payment amount - leakage

```python
#removing the columns stated above
col_rmv = ["id","member_id","funded_amnt","funded_amnt_inv","grade","sub_grade","emp_title","issue_d","zip_code","out_prncp","out_prncp_inv","total_pymnt","total_pymnt_inv","total_rec_prncp","total_rec_int","total_rec_late_fee","recoveries","collection_recovery_fee","last_pymnt_d","last_pymnt_amnt"]
loans_2007.drop(columns=col_rmv,axis=1,inplace=True)
print(loans_2007.shape)
loans_2007.columns
```

```python
loans_2007["loan_status"].value_counts()
```

Based on the given unique values, we would need to convert these things to a binary value of loan/no loan given. From the unique values there are 3 "major" areas:
1. loans that are fully paid (1)
2. loans that are charged off (0) - no longer expecting payment
3. Others.

We'll exclude the option #3, others - because of 2 issues: 
1. It's not reasonable to expect a machine learning algorithm to predict anything with substantially small data.
2. If the borrowers don't qualify for today (which we're trying to mimmic), we can't loan them out anyways.

```python
loans_2007 = loans_2007[(loans_2007["loan_status"] == "Fully Paid") | (loans_2007["loan_status"] == "Charged Off")].copy()
print(loans_2007.loan_status.value_counts())

#converting labels to either 1/0 (good/bad). Throw away option 3.
def target_conv(row):
    if row == "Fully Paid":
        return 1
    return 0

loans_2007["loan_status"] = loans_2007["loan_status"].apply(target_conv)
print(loans_2007.loan_status.value_counts())
```

```python
drop_cols = list()
cols = loans_2007.columns

for i in cols:
    #single value attribute won't contribute into our learning/model.
    temp = loans_2007[i].dropna().unique()
    if len(temp) <= 1: 
        print(temp,i)
        drop_cols.append(i)
drop_cols
```

```python
loans_2007.drop(drop_cols,axis=1,inplace=True)
loans_2007.columns
```

```python
loans_2007.isnull().sum()
```

It is quite important to have a known employment length as our parameter (if someone's unemployed, it defeats our purpose of lending the money). Better be safe than sorry. We'll remove the **column** "pub_rec_bankruptcies" and **rows** of missing data from the "title","revol_util", and "last_credit_pull_d"

```python
loans_2007.drop(["pub_rec_bankruptcies"],axis=1,inplace=True)
loans_2007.dropna(axis = 0,inplace=True)
print(loans_2007.shape)
loans_2007.isnull().sum()
```

```python
#only select object dtypes columns
loans_obj = loans_2007.select_dtypes(include=['object'])
loans_obj.dtypes
```

```python
loans_obj.head(10)
```

It seems that only "int_rate" and "revol_util" are continuous, others look like categorical. Let's confirm that with unique() method. The dates related data such as "earliest_cr_line" and "last_credit_pull_d" are not practical to be used with - due to decades long range, this might give a curse of dimensionality issue, thus needing more data.

```python
loans_obj.drop(["earliest_cr_line","last_credit_pull_d"],axis=1,inplace=True)
for col in loans_obj.columns:
    if (col != "int_rate") and (col != "revol_util"):
        print(loans_obj[col].unique())
```

```python
print(loans_obj.columns)
print(loans_obj["purpose"].value_counts()) #can we categorize this? If the "purpose" column doesn't have too many categories.
print(loans_obj["title"].value_counts()) #same as above
```

```python
#mapping of employment length to turn this into numerical
mapping = {
    "emp_length": {
        "10+ years": 10,
        "9 years": 9,
        "8 years": 8,
        "7 years": 7,
        "6 years": 6,
        "5 years": 5,
        "4 years": 4,
        "3 years": 3,
        "2 years": 2,
        "1 year": 1,
        "< 1 year": 0,
        "n/a": 0
    }
}
loans_2007.replace(mapping,inplace=True)

#making dummy variables for these columns below would prove to make our df too big to handle
#we'll keep the purpose column since it's manageable for our dummy variables
loans_2007.drop(columns=["last_credit_pull_d","addr_state","title","earliest_cr_line"],inplace=True) 

#converting both "int_rate" and "revol_util" into float
loans_2007['int_rate'] = loans_2007['int_rate'].str.rstrip('%')
loans_2007['int_rate'] = loans_2007['int_rate'].astype('float')
#print(loans['int_rate'].head())

loans_2007['revol_util'] = loans_2007['revol_util'].str.rstrip('%')
loans_2007['revol_util'] = loans_2007['revol_util'].astype('float')
#print(loans['revol_util'].head())
```

```python
dummy_df = pd.get_dummies(loans_2007[['home_ownership','verification_status', 'purpose', 'term']])#2dim

loans_2007 = pd.concat([loans_2007,dummy_df],axis=1)
#drop the original cols
loans_2007.drop(['home_ownership','verification_status', 'purpose', 'term'],axis=1,inplace=True)
```
