## Background & Objectives

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

path = "AB_NYC_2019.csv"
df = pd.read_csv(path)
print(df.dtypes)
df.head(5)
```

**Background:**

This data is provided by Airbnb on the year of 2019, in New York City. Since Airbnb is a successful, data driven company within the hospitality sector, we would can expect a lot of interesting finds within this dataset. Below are what we'll be doing throughout this notebook:


1. EDA on the given dataset to see any interesting finds.
2. Analyse various Machine Learning models to predict the prices of the listing.

Columns:

*ID*: Listing ID.

*Name*: Name of the listing.

*host_id*: Host ID.

*host_name*: Name of the host.

*neighborhood_groups* : General location.

*neighborhood* : The area within the neighborhood group.

*latitude* : Latitude.

*longitude* : Longitude.

*room_type*: Type of rooms (ie: Private Room, Entire home/apt).

*Price* : price per night.

*minimum_nights* : amount of days out of the year when listing is available for booking.

*number_of_reviews* : Total number of reviews.

*last_review* : Date of the last review.

*reviews_per_month* : Frequency of reviews per month.

*calculated_host_listings_count* : How many properties this host has.

*availability_365* : It is an amount of days out of the year when listing is available for booking.

## Data Wrangling

```python
#checking on missing data
print(df.shape)
print(df.isnull().sum())
```

```python
#since of of our misssing data are in 2 columns: last_review and reviews_per_month
#we're checking the data to see if there's anything special.
df[df["last_review"].isnull()].head(5)
```

```python
#does null values in columns: last_review and reviews_per_month indicate that there has never been any review ? 
#Indeed. These places simply has never gotten a review as shown below.
df[df["last_review"].isnull() & df["reviews_per_month"].isnull()]["number_of_reviews"].sum()
```

```python
#since these places have never gotten a review, we should then replace null values in the
#last_review, reviews_per_month as simply, 0.
df.fillna({'reviews_per_month':0,"last_review":0}, inplace=True)
df.isnull().sum()
```

```python
#now we can just simply drop the missing data from name, host_name (37 rows total). 
#Due to the size of our dataset, these won't pose a problem and/or provide extra insights
df.dropna(inplace=True)
print(df.shape)

#checking on prices column (where the prices are unreasonable, ie: $0 a night)
df[df["price"] <= 0]
```

```python
#remove those invalid prices, serves no purpose in prediction and/or analysis.
df = df[df["price"] > 0]

#using a new df for visualization purposes
vis_df = df.copy()
```

## Feature Engineering

Our goal is to help predict/suggest the price of the new listing by when a new host create a post. In order to do this, we would have to use regression models as well as feature engineering some of our existing features. 

Columns to drop:

Group1. id, name host_id, host_name, last_review: these conveys no numerical purposes

Group2. reviews_per_month, number_of_reviews: These are numerical and important, however they are a leakage since we need to predict the price based on a NEW host listing.
we'll keep the neighborhood_group and room types because we know that certain neighborhood_groups are more expensive than others (at the map data), and same with room_type.

```python
#df (not vis_df) is already cleaned with null values as above.
#these are columns we'll be using for prediction.
df_reg = df[["neighbourhood_group","room_type","minimum_nights","calculated_host_listings_count",
             "availability_365","latitude","longitude","price"]]
df_reg.head()
```

```python
#looking at average listing price for each of the neighbourhood_group in order to ordinal rank them.
#seems that Bronx is the worst, Staten Island and Queens are roughly similar. This was also indicated on one of the barplots above.
print(df_reg[df_reg["neighbourhood_group"] == "Bronx"]["price"].median())
print(df_reg[df_reg["neighbourhood_group"] == "Staten Island"]["price"].median())
print(df_reg[df_reg["neighbourhood_group"] == "Queens"]["price"].median())
print(df_reg[df_reg["neighbourhood_group"] == "Brooklyn"]["price"].median())
print(df_reg[df_reg["neighbourhood_group"] == "Manhattan"]["price"].median())
```

```python
print(df_reg["room_type"].value_counts())
print(df_reg["neighbourhood_group"].value_counts())
```

```python
#Generic labelencoder helper function
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

#either onehot encoding or ordinal encoding. Labelencoding doesn't help rank the neighbourhood_group and room_type for the price prediction.
#cover for both columns: "neighbourhood_group" and "room_type".
def ordinal_encoding(row):
    if row == "Bronx":
        return 1
    elif row == "Queens" or row == "Staten Island":
        return 2
    elif row == "Brooklyn":
        return 3
    elif row == "Manhattan":
        return 4
    
    elif row == "Shared room":
        return 1
    elif row == "Private room":
        return 2
    elif row == "Entire home/apt":
        return 3
    

#ordinal value ordered by usage of cost of rent in each neighbourhood group and the general "space" of the rental.
df_reg["neighbourhood_group"] = df_reg["neighbourhood_group"].apply(ordinal_encoding)
df_reg["room_type"] = df_reg["room_type"].apply(ordinal_encoding)
```

```python
#double checking our transformation/encoding
print(df_reg["room_type"].value_counts())
print(df_reg["neighbourhood_group"].value_counts())
```

## Machine Learning Models

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, SGDRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn import tree
from sklearn.model_selection import GridSearchCV

from keras import layers, Input, Model
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Activation
from keras.layers import BatchNormalization
from keras import optimizers
from keras import callbacks

#new
from keras.utils import plot_model
```

```python
class modeling:
    
    def __init__(self,df):
        self.df = df #input df
        self.X_train = None 
        self.X_test = None
        self.y_train = None 
        self.y_test = None
        
        #scaler for standard scaling and minmax scaling.
        self.X_scaler = None
        self.X_train_ss = None
        self.X_test_ss = None
        
        #mainly for neural network for convergence.
        self.y_scaler = 500.0
        self.y_train_ss = None #for neural network.
        self.y_test_ss = None #for neural network.
        
        #df to display the training/testing results based off model list and index list
        self.result_table = None
        
        self.model_list = []
        self.index_list = []
        
        #store predicted results
        self.nn_model = None
        self.history = None
    
    
    def train_test_split(self,ratio=0.25,seed=42):
        '''
        function: to split train/test
        Parameters:
            ratio of split, seed number.
        Return:
            set the training/testing attributes.
        '''
        assert(self.df is not None), "no dataframe to split train/test"
        try:    
            X = self.df.drop(columns=["price"])
            y = self.df["price"]
            self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(X,y,test_size=ratio,random_state=seed)
        
        except:
            print("check feature columns and label 'price' column to make sure df is functional")
    
    
    def standard_scaling(self):
        '''
        Function: This function will use Standard Scaling on training set, and using the fitting on the whole training dataset, setting them into the scaled variables
            of the attribute list.
        Parameters:
            None
        Return:
            None.
        '''
        features = X_train.columns.tolist()
        assert((self.X_train is not None) and (self.X_test is not None) and (self.y_train is not None) and (self.y_test is not None)),"call method self.train_test_split() prior"
        assert(features),"cannot feature scaling with no feature columns. Check main df"
        
        standard_scaler = preprocessing.StandardScaler(with_mean=True,with_std=True)
        self.X_scaler = standard_scaler.fit(self.X_train)
        self.X_train_ss = standard_scaler.transform(self.X_train)
        self.X_test_ss = standard_scaler.transform(self.X_test)
        
        
        #convert the newly scaled data into dataframes
        self.X_train_ss = pd.DataFrame(self.X_train_ss,columns=features)
        self.X_test_ss = pd.DataFrame(self.X_test_ss,columns=features)
        
        #since we truncated at $500, for neural network usage
        self.y_train_ss = self.y_train/500.0
        self.y_test_ss = self.y_test/500.0
    
    def minmax_scaling(self):
        '''
        Function: This function will use MinMax Scaling on training set, and using the fitting on the whole training dataset, setting them into the scaled variables
            of the attribute list.
        Parameters:
            None
        Return:
            None.
        '''        
        features = self.X_train.columns.tolist()
        assert((self.X_train is not None) and (self.X_test is not None) and (self.y_train is not None) and (self.y_test is not None)),"call method self.train_test_split() prior"
        assert(features),"cannot feature scaling with no feature columns. Check main df"
        
        minmax_scaler = preprocessing.MinMaxScaler()
        self.X_scaler = minmax_scaler.fit(self.X_train)
        self.X_train_ss = minmax_scaler.transform(self.X_train)
        self.X_test_ss = minmax_scaler.transform(self.X_test)
        
        #convert the newly scaled data into dataframes
        self.X_train_ss = pd.DataFrame(self.X_train_ss,columns=features)
        self.X_test_ss = pd.DataFrame(self.X_test_ss,columns=features)
        
        #since we truncated at $500, for neural network usage
        self.y_train_ss = self.y_train/500.0
        self.y_test_ss = self.y_test/500.0
    
    def summary(self):
        '''
        fucntion: to establish a df for viewing the result
        Parameters: 
            model_lst: list of dictionary of results
            index: list of the names of the models in that order

        Return:
            A df with model training and testing results
        '''
        assert (len(self.model_list) > 0) and (len(self.index_list)==len(self.model_list)),"empty model_list attribute or mismatched sizes of model_list/index_list attributes"
        
        df = pd.DataFrame(columns=["train_mse","train_mae","test_mse","test_mae"])
        for i in range(len(self.model_list)):
            temp = pd.DataFrame(self.model_list[i],[self.index_list[i]]) #values must be in a 2d list for df[[]]
            df = df.append(temp)
        self.result_table = df
        print(self.result_table)
        
        
    def add_model(self,model,name,scaled=0):
        '''
        function: to add a model for evaluation.
        Parameters:
            model: sklearn model
            name: key value of the added model.
            scale: using the scaled data (need to call self.standard_scaling() or self.minmax_scaling() methods first)
        Return:
            None
        '''
        assert(name) not in self.index_list,"name parameter must be changed since it's already existed in the result table"
        if not scaled:
            m = model
            m.fit(self.X_train,self.y_train)
            m_pred = m.predict(self.X_test)
            
            train_mse = mean_squared_error(m.predict(self.X_train),self.y_train)
            train_mae = mean_absolute_error(m.predict(self.X_train),self.y_train)

            test_mse = mean_squared_error(m_pred,self.y_test)
            test_mae = mean_absolute_error(m_pred,self.y_test)

            #return in dict format for later usage
            self.model_list.append({"train_mse":[train_mse], "train_mae":[train_mae], "test_mse":[test_mse], "test_mae":[test_mae]})
            self.index_list.append(name)
            
        elif scaled:
            #unless using NN, scaling the target is not needed
            assert(self.X_scaler), "need to scale the data prior to modeling"
            m = model
            m.fit(self.X_train_ss,self.y_train)
            m_pred = m.predict(self.X_test_ss)

            train_mse = mean_squared_error(m.predict(self.X_train_ss),self.y_train)
            train_mae = mean_absolute_error(m.predict(self.X_train_ss),self.y_train)
            test_mse = mean_squared_error(m_pred,self.y_test)
            test_mae = mean_absolute_error(m_pred,self.y_test)

            #return in dict format for later usage
            self.model_list.append({"train_mse":[train_mse], "train_mae":[train_mae], "test_mse":[test_mse], "test_mae":[test_mae]})
            self.index_list.append(name)
        
        
    def build_nn(self,depth_matrix):
        '''
        function: to add a model for evaluation. Set self.nn_model after the build.
        Parameters:
            depth_matrix: list of nodes for each hidden layer. Ie: [10,10,10] for 3 hidden layers of 10 nodes each.
        Return:
            None
        '''
        input_tensor = Input(shape=(self.X_train.shape[1],))
                             
        x = Dense(depth_matrix[0])(input_tensor)
        for i in range(1,len(depth_matrix)):
            x = Dense(depth_matrix[i])(x)
            x = Activation("relu")(x)
            x = Dropout(rate=0.5)(x)
            x = BatchNormalization()(x)
            
        output_tensor = Dense(1)(x)
        model = Model(inputs=input_tensor,outputs=output_tensor)
        model.compile(optimizer=optimizers.Adam(lr=0.00005),loss="mse",metrics=["mae","mse"])
        
        print(model.summary())
        #graphviz needed
        #plot_model(model,show_shapes=True)
        self.nn_model = model
    
    def plot_history(self):
        '''
        function: helper function that will output training/validation plots for neural network after training.
        Parameter:
            None
        Return:
            None
        '''
        fig,(ax1,ax2) = plt.subplots(1, 2,figsize=(30,10))
        
        #we multiply the mae/mse by $500 because that's what we rescaled our target down to from the original 0-500 dollars range target.
        ax1.plot([i for i in range(1,len(self.history.history["mae"])+1)],
                 self.history.history["mae"])
        ax1.plot([i for i in range(1,len(self.history.history["val_mae"])+1)],
                 self.history.history["val_mae"])
        ax1.legend(["train_mae","val_mae"])
        ax1.set_title("MAE")
        ax1.set_ylim(0,0.2)
        
        ax2.plot([i for i in range(1,len(self.history.history["mse"])+1)],
                 self.history.history["mse"])
        ax2.plot([i for i in range(1,len(self.history.history["val_mse"])+1)],
                 self.history.history["val_mse"])
        ax2.legend(["train_mse","val_mse"])
        ax2.set_title("MSE")    
        ax2.set_ylim(0,0.2)
    
    def nn_train(self,ep,b_size,scaling="standard",plotting=True):
        '''
        function: to train neutral network after building it. Need to call self.build_nn() prior.
        Parameter:
            ep: number of epochs
            b_size: batch size
            scaling: either "standard" or "minmax" for scaling the data prior to training. Default is "standard".
            plotting: to call self.plot_history() function to plot the training/validation data. Default is True.
        '''
        assert(self.nn_model is not None),"Need to build model first before training"
        
        
        if scaling =="standard":
            self.standard_scaling()
        elif scaling == "minmax":
            self.minmax_scaling()
        else:
            raise("scaling parameter needs to be either 'standard' or 'minmax' ")
        self.history = self.nn_model.fit(self.X_train_ss,self.y_train_ss,epochs=ep,batch_size=b_size,
                                     validation_data=(self.X_test_ss,self.y_test_ss),verbose=True)
        if plotting:
            self.plot_history()
        
        
```

```python
#evaluating baseline models for screening.
model = modeling(df_reg)
model.train_test_split()
model.standard_scaling()
model.add_model(LinearRegression(),"lr_bl",scaled=0)
model.add_model(SGDRegressor(loss="huber"),"SGD_bl",scaled=1) #huber loss is mse but less sensitive with outliers - which are many of our high-end rent data
model.add_model(Ridge(),"ridge_bl",scaled=1)
model.add_model(Lasso(),"lasso_bl",scaled=1)
model.add_model(SVR(kernel='linear'),"SVRlinear_bl",scaled=1)
model.add_model(SVR(kernel='rbf'),"SVRrbf_bl",scaled=1)
model.add_model(tree.DecisionTreeRegressor(),"dt_bl",scaled=0)
model.add_model(tree.ExtraTreeRegressor(),"extratree_dl",scaled=0)
model.add_model(XGBRegressor(),"xgb_bl",scaled=0)
model.add_model(RandomForestRegressor(criterion="mae",n_jobs=-1),"rf_bl",scaled=0) #previous runs see that mae loss outperform mse due to many outliers that we have for target
model.summary()
```

Since our df_reg data has a price mean of 130 dollars and the standard deviation of 85 dollars, it seems that some of these baselines are all great as the MAEs are 1/2 a sigma away from the std. We can finetune these further to find the best model. Particularly - XGB, RF, SRVrbf, SGD are the ones that we should focus our energy on. This is expected as these algorithms are known to be more complex than simply LR. Ensemble and SVM are some of the top performers for shallow learning.

DT seem to be overfitting quite a bit, as expected. However, RF model overfitting is another big question because it is designed to lower the variance. Yet, it's still less overfitting compared to DT as expected.

```python
#inheriting modeling class
class gridsearch(modeling):
    def __init__(self,df):
        super().__init__(df)
        self.model_list = dict() #holding all gridsearched models and its results
        self.best_models = dict()
        
    
    def gridsearchcv(self,model,name,hyperparameters_grid,scale_type):
        '''
        function: to do gridsearch and store the results of our gridsearch in self.model_list dictionary.
        Parameters:
            model: a baseline model from sklearn.
            name: key name for the model being gridsearch to store in the dictionary.
            hyperparameters_grid: a dictionary of our DOE/gridsearch.
            scale_type: either "minmax" or "standard" for scaling the features
        Return:
            None
        '''
        
        #record 2 metrics, but we optimize with refit="MAE" metric
        grid = GridSearchCV(model,hyperparameters_grid,scoring=["neg_mean_absolute_error","neg_mean_squared_error"],refit="neg_mean_absolute_error")
        if not scale_type:
            grid.fit(self.X_train,self.y_train)
            
        elif scale_type == "minmax":
            self.train_test_split()
            self.minmax_scaling()
            grid.fit(self.X_train_ss,self.y_train_ss)
        
        elif scale_type == "standard":
            self.train_test_split()
            self.standard_scaling()
            grid.fit(self.X_train_ss,self.y_train_ss)
            
        self.model_list[name] = grid.cv_results_
        
        
    def extract_best_params(self):
        '''
        function: to extract out the best model of gridsearch models and store them in the dictionary of self.best_models.
        Parameters:
            None
        Return:
            None
        '''
        assert(bool(self.model_list)),"no grid search was done to extract out best parameters. Need to do gridsearch first"
        for model in self.model_list:
            rank_list = self.model_list[str(model)]["rank_test_neg_mean_absolute_error"]
            index = rank_list.tolist().index(1) #return the #1 best set of hyperparameter ranked of the list.
            best_params = self.model_list[str(model)]["params"][index] #extract out the best hyperparameters of our gridsearch
            self.best_models[model] = best_params
    
    
            
        
        
```
