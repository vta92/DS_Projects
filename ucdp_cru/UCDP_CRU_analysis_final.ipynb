Question:

```python
from netCDF4 import Dataset
import xarray as xr
```

```python
#better to sort the files chronologically. Help with stacking the DF's.
path_avg = ['cru_ts4.04.1981.1990.tmx.dat.nc','cru_ts4.04.1991.2000.tmx.dat.nc',
         'cru_ts4.04.2001.2010.tmx.dat.nc','cru_ts4.04.2011.2019.tmx.dat.nc']
path_min = ['cru_ts4.04.1981.1990.tmn.dat.nc','cru_ts4.04.1991.2000.tmn.dat.nc',
           'cru_ts4.04.2001.2010.tmn.dat.nc','cru_ts4.04.2011.2019.tmn.dat.nc']
path_max = ['cru_ts4.04.1981.1990.tmp.dat.nc','cru_ts4.04.1991.2000.tmp.dat.nc',
           'cru_ts4.04.2001.2010.tmp.dat.nc','cru_ts4.04.2011.2019.tmp.dat.nc']
path_pre = ['cru_ts4.04.1981.1990.pre.dat.nc','cru_ts4.04.1991.2000.pre.dat.nc',
           'cru_ts4.04.2001.2010.pre.dat.nc','cru_ts4.04.2011.2019.pre.dat.nc']
```

```python

def open_cru(path):
    '''
    function: to stack the CRU df's chronologically.
    Parameter: a list of filepath
    Return: A stacked df based on the order of input.
    '''
    assert (path is not None), "list to files path is empty for processing"
    
    df = pd.DataFrame()
    
    for p in path:
        f = xr.open_dataset("cru/{0}".format(p))
        temp = f.to_dataframe()
        #dropping 'stn' column
        temp.drop("stn",axis=1,inplace=True)
        frames = [df,temp]
        df = pd.concat(frames)
        
    #df = df.reset_index() #seems faster to do outter join with index later on prior to reseting.
    #df["time"] = pd.to_datetime(df["time"]) #time is part of the index
        
    return df
```

```python
import time

a = time.time()
df_avg = open_cru(path_avg)
df_min = open_cru(path_min)
df_max = open_cru(path_max)
df_pre = open_cru(path_pre)

print(time.time() - a)
```

```python
print(df_avg.isnull().sum())
print(df_min.isnull().sum())
print(df_max.isnull().sum())
print(df_pre.isnull().sum())
print(df_avg.shape)
```

```python
#checking the null time/coordinates. It seems that they are just there in general throughout the time and along the needed latitudes
df_avg[df_avg.isnull()].index
```

```python
#we're joining tables with multilevel indices (lat,lon,date)
a = time.time()
df_cru_final = pd.concat([df_avg,df_min,df_max,df_pre],axis=1)
print(time.time() - a)
```

```python
#this proves that the missing data are consistent in time/latitude where measurements did not happen those days.
#hopefully these are somewhere in the ocean
df_cru_final.isnull().sum()
```

```python
#reseting indices as 'lat','lon',time' columns
df_cru_final.reset_index(inplace=True)
df_cru_final["time"] = pd.to_datetime(df_cru_final["time"])
#creating month/year column for grouping
df_cru_final["month"] = df_cru_final["time"].dt.month
df_cru_final["year"] = df_cru_final["time"].dt.year
df_cru_final.head()
```

```python
#We'll assume that the general coordinates for these places can act as a reference for the whole region within each:
#(Kunduz, Ghazni, and Kabul) districts. Down to 2 sig digits from from the previous map plot.
#Kunduz: (36.5 to 37.5), (68.5 to 69.5)
#Ghazni: (32.5 to 33.5), (67.5 to 68.5)
#Kabul: (34.3 to 35), (69 to 70)
mask_kunduz = (df_cru_final["lat"] > 36.5)&(df_cru_final["lat"] < 37.5)&(df_cru_final["lon"] > 68.5)&(df_cru_final["lon"] < 69.5)
mask_ghazni = (df_cru_final["lat"] > 32.5)&(df_cru_final["lat"] < 33.5)&(df_cru_final["lon"] > 67.5)&(df_cru_final["lon"] < 68.5)
mask_kabul = (df_cru_final["lat"] > 34.3)&(df_cru_final["lat"] < 35)&(df_cru_final["lon"] > 69)&(df_cru_final["lon"] < 70)

df_cru_kunduz = df_cru_final[mask_kunduz]
df_cru_ghazni = df_cru_final[mask_ghazni]
df_cru_kabul = df_cru_final[mask_kabul]
```

# Question 3:

We'll be making use of seasons for better visualization and test in correlation table to show if higher temperature or precipitation have any correlation to the total affected tolls. Benchmark: all 3 cities MUST have abs(R) > 2 with similar trends/R-values.


Seasons defined:

1 = Spring

2 = Summer

3 = Fall

4 = Winder

```python
#from the previous plot, we divide up the seasons.
def seasons(month):
    month = int(month)
    if (month >= 3) and  (month < 6):
        return 1
    elif (month >= 6) and (month < 9):
        return 2
    elif (month >= 9) and (month < 12):
        return 3
    elif (month == 12) or (month < 3):
        return 4
    else:
        return -1
df_combined["season"] = df_combined["month"].apply(seasons)
```

# All 3 districts temp/precip vs. conflict tolls

```python
ax1 = df_combined[["year","season","tmx","tmp","tmn"]].groupby(by=["season","year"]).mean().plot.line(title="Temperature for All 3 districts with Lag of 1 Year")
ax2 = df_combined[["year","season","best"]].groupby(by=["season","year"]).mean().plot.line(title="Death Tolls for All 3 districts with Lag of 1 Year")
ax3 = df_combined[["year","season","pre"]].groupby(by=["season","year"]).mean().plot.line(title="Precipitation for All 3 districts with Lag of 1 Year")
df_combined[["year","season","tmx","tmp","tmn","pre","best"]].groupby(by=["season","year"]).mean().corr()

ax1.set_ylabel("Temeperature (C)")
ax2.set_ylabel("Total Death Tolls")
ax3.set_ylabel("Precip. (mm/day)")
```

These graphs are divided by (seasons,year) generally for better understanding of climate change. In which, season 1 is spring and season 4 is winter. As we can see from plot #1, it seems that year over year for the past 40 years, the summers and winters have been getting a little hotter (season=2,4). The variation in temperature between the years are less than the temperature variation within a single season. This shows by the lack of trendline within each season as the years progress. The variation within each season (noise) is a much bigger factor over the years. Perhaps the variation itself is a result of global warming (as cited in many papers). However, there doesn't show any clear indicator between the temperature and the conflict tolls generally.
